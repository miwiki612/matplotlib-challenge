### Analysis

- Data Cleaning

There are a total of 249 mice in this study. Mouse “g989” has multiple distinct entries at the same Timepoint, since we are unable to confirm which entry is right, Mouse “g989” is removed from the result. 248 mice are included in this analysis.

- Potential Treatment

Based on our Summary, there are 5 treatments that show better performance: Capomulin, Ceftamin, Infubinol, Propriva, Ramicane. Average of the Tumor Volume and Median Tumor Volume in these 5 treatments is less than others. But Propriva has significantly higher Variance. Therefore we did detailed analysis on 4 treatments: Capomulin, Ceftamin, Infubinol, Ramicane.

For 4 selected treatments, box plot shows mouses with Capomulin and Ramicane treatment has smaller Final Tumor Volume. Infubinol has an outlier that shows significantly smaller Final Tumor Volume.

- Concerns

However the analysis on mouse Weight (g) and Tumor Volume shows high correlation. The Tumor Volume can be impacted by variance on mouse’s Weight each treatment had. To have an effective analysis on treatment, we need to avoid the impact on mouse Weight. The study should be analyzed with standardized change on Tumor Volume rather than absolute value. Ideally the mouse Weight for all treatments should be standardized or have a similar structure.

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 

drug_group = study_df.groupby( 'Drug Regimen' )

# mean, median, variance, standard deviation, and SEM of the tumor volume. 

drug_group_mean = drug_group.mean( numeric_only=True )['Tumor Volume (mm3)'].to_frame( name = "Mean Tumor Volume" )
drug_group_median = drug_group.median( numeric_only=True )['Tumor Volume (mm3)'].to_frame( name = "Median Tumor Volume" )
drug_group_var = drug_group.var( numeric_only=True )['Tumor Volume (mm3)'].to_frame( name = "Tumor Volume Variance" )
drug_group_sd = drug_group.std( numeric_only=True )['Tumor Volume (mm3)'].to_frame( name = "Tumor Volume Std. Dev" )
drug_group_sem = drug_group.sem( numeric_only=True )['Tumor Volume (mm3)'].to_frame( name = "Tumor Volume Std. Err." )

# Assemble the resulting series into a single summary DataFrame.
# df_2 compare to aggregate method, the not prefered is setted as df_2
drug_tumor_volume_df_2 = pd.merge( drug_group_mean , drug_group_median , how = 'outer' , on = 'Drug Regimen')
drug_tumor_volume_df_2 = pd.merge( drug_tumor_volume_df_2 , drug_group_var , how = 'outer' , on = 'Drug Regimen')
drug_tumor_volume_df_2 = pd.merge( drug_tumor_volume_df_2 , drug_group_sd , how = 'outer' , on = 'Drug Regimen')
drug_tumor_volume_df_2 = pd.merge( drug_tumor_volume_df_2 , drug_group_sem , how = 'outer' , on = 'Drug Regimen')

# display
drug_tumor_volume_df_2
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
# Using the aggregation method, produce the same summary statistics in a single line

drug_tumor_volume_df = study_df.groupby( 'Drug Regimen' )[ 'Tumor Volume (mm3)'].aggregate( [ 'mean' , 'median' , 'var' , 'std' , 'sem' ])

# use first method column names to reset colnames
drug_tumor_volume_df.columns = drug_tumor_volume_df_2.columns

# display
drug_tumor_volume_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:

# Capomulin, Ramicane, Infubinol, and Ceftamin
sub_drug_list = [ 'Capomulin' , 'Ramicane' , 'Infubinol' , 'Ceftamin' ]
sub_study_df = study_df[ study_df['Drug Regimen'].isin(sub_drug_list) ]

# Start by getting the last (greatest) timepoint for each mouse
last_point = sub_study_df.groupby('Mouse ID' ).max()['Timepoint'].to_frame().reset_index()
last_point['KEEP'] = True

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
sub_study_df = pd.merge( sub_study_df , last_point, on=["Mouse ID","Timepoint"] , how='outer')

# drop KEEP = NaN, drop column KEEP
sub_study_df = sub_study_df[ sub_study_df['KEEP'].notna() ]
sub_study_df = sub_study_df.drop( columns='KEEP' )

print( sub_study_df.shape )
print( sub_study_df['Drug Regimen'].value_counts() )
sub_study_df.head()
```

```python
# ---------- instuction
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
# Calculate the IQR and quantitatively determine if there are any potential outliers.

# ---------- solution
# Method : create disctionary then transpose the dataframe to manageable structure
# disctionary structure = { 'drug' : [ 'min' , 'Q1', 'median' , 'Q3', 'max' , 'iqr' , 'lower_bound' , 'upper_bound' ] }

# initial keys
sub_drug_vol_box_df = { drug : None for drug in sub_drug_list }

# collect quantile data for each drug
for drug in sub_drug_list :
    
    # data for one drug
    temp_drug_df = sub_study_df[ sub_study_df['Drug Regimen'] == drug ]
    temp_drug_df = temp_drug_df[ 'Tumor Volume (mm3)' ]
    
    # list for min, 0.25, 0.5 , 0.75, max , iqr
    temp_list = []
    
    temp_list.append( temp_drug_df.min() )
    
    temp_quantile = temp_drug_df.quantile([.25,.5,.75])
    temp_list.extend( temp_quantile )
    
    temp_list.append( temp_drug_df.max() )
    
    # iqr
    temp_iqr = temp_quantile[0.75] - temp_quantile[0.25]
    temp_list.append( temp_iqr )
    
    # lower_bound
    temp_list.append( temp_quantile[0.25] - 1.5 * temp_iqr )
    # upper_bound
    temp_list.append( temp_quantile[0.75] + 1.5 * temp_iqr )
    
    # add data to this drug
    sub_drug_vol_box_df[drug] = temp_list
```

```python
# Transfer dictionary to df then transpose df, then rename column
sub_drug_vol_box_df = pd.DataFrame( sub_drug_vol_box_df ).T
sub_drug_vol_box_df.columns = [ 'min' , 'Q1', 'median' , 'Q3', 'max' , 'iqr' , 'lower_bound' , 'upper_bound' ]
sub_drug_vol_box_df
```

```python
# ---------- instuction
# Locate the rows which contain mice on each drug and get the tumor volumes    
# add subset
# Determine outliers using upper and lower bounds

# ---------- solution
# Method : use each drug's upper and lower bound to select this drug's outlier then display the result.

for i in range(len(sub_drug_vol_box_df)) :
    temp_drug_df = sub_study_df[ sub_study_df['Drug Regimen'] == sub_drug_vol_box_df.index[i] ]
    temp_drug_df = temp_drug_df[(temp_drug_df['Tumor Volume (mm3)'] > sub_drug_vol_box_df['upper_bound'][i]) | \
                                (temp_drug_df['Tumor Volume (mm3)'] < sub_drug_vol_box_df['lower_bound'][i]) ]
        
    # print result for subset contains outliers
    
    # when this drug has no outlier
    if len(temp_drug_df) == 0:
        print(f"{sub_drug_vol_box_df.index[i]} doesn't have potential outlier.")
    
    # when this drug contains outlier
    else :
        print(f"{sub_drug_vol_box_df.index[i]} has {len(temp_drug_df)} potential outliers:")
        print("----------")
        for j in range(len(temp_drug_df)) :
            print("Mouse ID:"+temp_drug_df.iloc[j]['Mouse ID']+". Timepoint:"+str(temp_drug_df.iloc[j]['Timepoint'])+\
                  ". Final Tumor Volume (mm3):"+str(temp_drug_df.iloc[j]['Tumor Volume (mm3)']) )
        print("----------")        
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# ---------- solution
# Since Weight for each mouse doesn't change, the data frame for this question will be groupby Mouse ID and Weight

capomulin_df_avg = capomulin_df.groupby(['Mouse ID','Weight (g)']).mean(numeric_only=True)['Tumor Volume (mm3)'].reset_index()
fig, ax = plt.subplots()
ax.scatter( capomulin_df_avg['Weight (g)'] , capomulin_df_avg['Tumor Volume (mm3)'])
ax.set_xlabel( "Weight (g)")
ax.set_ylabel( "Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# linear regression model and line
x_values = capomulin_df_avg['Weight (g)']
y_values = capomulin_df_avg['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

# linear regression value
regress_values = x_values * slope + intercept

# linear regression equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# scatter plot
fig, ax = plt.subplots()
ax.scatter( capomulin_df_avg['Weight (g)'] , capomulin_df_avg['Tumor Volume (mm3)'] )
ax.set_xlabel( 'Weight (g)' )
ax.set_ylabel( 'Average Tumor Volume (mm3)' )

# regression line and equation
ax.plot(x_values,regress_values,"r-")
ax.annotate(line_eq,(21,35),fontsize=15,color="red")

print(f"The correlation between mouse weight and the average tumor volume is {round(rvalue,2)}")

plt.show()  
```
